```python
import cv2
from google.colab.patches import cv2_imshow
import torch
import matplotlib.pyplot as plt
```

```python
# YOLO - You Only Look Once

yolo = torch.hub.load('ultralytics/yolov5', 'yolov5x', pretrained=True)
```

```python
img1 = cv2.imread('/content/Image.jpg')
cv2_imshow(img1)
```

```python
object_detection1 = yolo(img1)
```

```python
cv2_imshow(object_detection1.render()[0])
```

```python
img2 = cv2.imread('/content/Image 2.jpg')
cv2_imshow(img2)
```

```python
object_detection2 = yolo(img2)
```

```python
cv2_imshow(object_detection2.render()[0])
```
